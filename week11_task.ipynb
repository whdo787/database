```python
with con:
  data = con.execute("""
  SELECT Student_ID, Course_ID, SUM(Credits), SUM(GPA_Grade * credits) / SUM(4.5 * credits) AS GPA
FROM (
  SELECT Student_ID, Course_ID, Credits, 
    CASE
        WHEN Grade = 'A+' THEN 4.50
        WHEN Grade = 'A' THEN 4.00
        WHEN Grade = 'B+' THEN 3.50
        WHEN Grade = 'B' THEN 3.00
        WHEN Grade = 'C+' THEN 2.50
        WHEN Grade = 'C' THEN 2.00
        WHEN Grade = 'D+' THEN 1.50
        WHEN Grade = 'D' THEN 1.00
        WHEN Grade = 'F' THEN 0
        END AS GPA_Grade
  FROM Spring_Semester
  )
  GROUP BY Student_ID
  """)
  for row in data:
    print(row)
```
