```python
# Prints unique titles in CSV, case insensitively

import csv

# For accumulating (and later sorting) titles
titles = set()

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file, adding each (uppercased) title to set
    for row in reader:
        titles.add(row["title"].strip().upper())

# Print titles in sorted order
for title in sorted(titles):
    print(title)
```

```python
# Prints popularity of titles in CSV, sorted by title

import csv

# For accumulating (and later sorting) titles
titles = {}

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file, adding each (uppercased) title to dictionary
    for row in reader:

        # Canoncalize title
        title = row["title"].strip().upper()

        # Count title
        if title in titles:
            titles[title] += 1
        else:
            titles[title] = 1

# Print titles in sorted order
for title in sorted(titles):
    print(title, titles[title])
```

```python
# Prints popularity of titles in CSV, sorted by popularity

import csv

# For accumulating (and later sorting) titles
titles = {}

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file, adding each (uppercased) title to dictionary
    for row in reader:

        # Canoncalize title
        title = row["title"].strip().upper()

        # Count title
        if title in titles:
            titles[title] += 1
        else:
            titles[title] = 1

# Function for comparing titles by popularity
def get_value(title):
    return titles[title]

# Print titles in sorted order
for title in sorted(titles, key=get_value, reverse=True):
    print(title, titles[title])
```

```python
# Prints popularity of titles in CSV, sorted by popularity, using a lambda function

import csv

# For accumulating (and later sorting) titles
titles = {}

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file
    for row in reader:

        # Canoncalize title
        title = row["title"].strip().upper()

        # Update counter
        if title in titles:
            titles[title] += 1
        else:
            titles[title] = 1

# Print titles in sorted order
for title in sorted(titles, key=lambda title: titles[title], reverse=True):
    print(title, titles[title])
```

```python
import csv
import re

# For counting mentions of The Office
counter = 0

# Open CSV File
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file, counting mentions of OFFICE and THE OFFICE
    for row in reader:
        title = row["title"].strip().upper()
        if re.search("^(OFFICE|THE OFFICE)$", title):
            counter += 1

# Print popularity
print(f"Number of people who like The Office: {counter}")
```

```python
# Searches CSV for popularity of a title

import csv

# Prompt user for title
title = input("Title: ").strip().upper()

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)

    # Iterate over CSV file, counting favorites
    counter = 0
    for row in reader:
        if row["title"].strip().upper() == title:
            counter += 1

# Print popularity
print(counter)
```

```python
# Imports titles and genres from CSV into a SQLite database

import csv

# Create database
from sqlalchemy import create_engine
db = create_engine("sqlite:////content/favorites.db/")
print(db.name)


# Create tables
db.execute("CREATE TABLE shows (id INTEGER, title TEXT NOT NULL, PRIMARY KEY(id))")
db.execute("CREATE TABLE genres (show_id INTEGER, genre TEXT NOT NULL, FOREIGN KEY(show_id) REFERENCES shows(id))")

r_set=db.execute('''select name from sqlite_master where type = 'table' ''')
for row in r_set:
  print(row)

# Open CSV file
with open("favorites.csv", "r") as file:

    # Create DictReader
    reader = csv.DictReader(file)
    show_id = 1
    # Iterate over CSV file
    for row in reader:

        # Canoncalize title
        title = row["title"].strip().upper()

        # Insert title
        db.execute("INSERT INTO shows (id, title) VALUES(?,?)",show_id,title)
        show_id = show_id + 1

        # Insert genres
        for genre in row["genres"].split(", "):
            print(genre + "||"+ str(show_id))
            db.execute("INSERT INTO genres (show_id, genre) VALUES(?, ?)", show_id, genre)
            
```

```python
# Searches CSV for popularity of a title

import csv

# Open database
from sqlalchemy import create_engine
db = create_engine("sqlite:////content/favorites.db/")
print(db.name)

# Prompt user for title
title = input("Title: ").strip()

# Search for title
r_set = db.execute("SELECT COUNT(*) AS counter FROM shows WHERE title LIKE ?", "%" + title + "%")

# Get first (and only) row
# Print popularity
data=r_set.fetchall()
for row in data:
    print(row[0])

```
